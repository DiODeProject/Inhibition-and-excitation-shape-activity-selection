## This notebook may be used to reproduce Fig S1 in the Supplementary Material

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import os
```

```python
fileDir = os.path.dirname(os.path.realpath('./DataGeneration/DataFiles/__file__'))

DecTimeFilePooled = []
ValueFilePooled = []
Option2RatePooled = []
UndecidedRatePooled = []

filenameInterneuronal = os.path.join(fileDir, 'Data2018_DT_ER_V_PooledEqualAltAbsB_randSeed.csv')
#filenameInterneuronal = os.path.join(fileDir, 'Data2018_DT_ER_V_PooledEqualAltAbsB_negMotivationsAllowed.csv')
with open(filenameInterneuronal, 'r') as infile:
    infile.readline
    for line in infile:
        if not line.startswith('#'): 
            ValueFilePooled.append(float(line.split(',')[0]))  
            DecTimeFilePooled.append(float(line.split(',')[1]))
            Option2RatePooled.append(float(line.split(',')[3]))  
            UndecidedRatePooled.append(float(line.split(',')[4]))

DecTimeFilePooled2 = []
ValueFilePooled2 = []
Option2RatePooled2 = []
UndecidedRatePooled2 = []
filenameInterneuronal = os.path.join(fileDir, 'Data2018_DT_ER_V_PooledUnequalAltDiffAbsB_randSeed.csv')
#filenameInterneuronal = os.path.join(fileDir, 'Data2018_DT_ER_V_PooledUnequalAltDiffAbsB_negMotivationsAllowed.csv')
with open(filenameInterneuronal, 'r') as infile:
    infile.readline
    for line in infile:
        if not line.startswith('#'): 
            ValueFilePooled2.append(float(line.split(',')[0]))  
            DecTimeFilePooled2.append(float(line.split(',')[1]))
            Option2RatePooled2.append(float(line.split(',')[3]))  
            UndecidedRatePooled2.append(float(line.split(',')[4]))

DecTimeFilePooled3 = []
ValueFilePooled3 = []
Option2RatePooled3 = []
UndecidedRatePooled3 = []
filenameInterneuronal = os.path.join(fileDir, 'Data2018_DT_ER_V_PooledUnequalAltRatioAbsB_randSeed.csv')
#filenameInterneuronal = os.path.join(fileDir, 'Data2018_DT_ER_V_PooledUnequalAltRatioAbsB_negMotivationsAllowed.csv')
with open(filenameInterneuronal, 'r') as infile:
    infile.readline
    for line in infile:
        if not line.startswith('#'): 
            ValueFilePooled3.append(float(line.split(',')[0]))  
            DecTimeFilePooled3.append(float(line.split(',')[1]))
            Option2RatePooled3.append(float(line.split(',')[3]))  
            UndecidedRatePooled3.append(float(line.split(',')[4]))
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.axes()
plt.plot(ValueFilePooled, DecTimeFilePooled, 'or', markersize=14, 
         ls = 'solid', lw = 3.5, label=r'$\Delta d = 0$')
plt.plot(ValueFilePooled2, DecTimeFilePooled2, 'sg', markersize=14, 
         ls = 'solid', lw = 3.5, label=r'$\Delta d = 2$')
plt.plot(ValueFilePooled3, DecTimeFilePooled3, '^b', markersize=14, 
         ls = 'solid', lw = 3.5, label=r'$\rho_d = 2$')


#plt.plot(Val1, RT_mean_ex, 'hb', markersize=15, lw = 3.5)
plt.xlabel(r'mean deficit $d_m$', fontsize=28)
plt.ylabel(r'mean decision time $\langle \mathsf{DT} \rangle$', fontsize=28)

#plt.xlim(0,1)
#plt.ylim(0.4,2.5)
#plt.ylim(0.6,0.9)

ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.25))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.1))
#ax.yaxis.set_major_locator(ticker.MultipleLocator(0.1))
#ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.02))
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
    
for tick in ax.xaxis.get_major_ticks():
                tick.label.set_fontsize(22) 
for tick in ax.yaxis.get_major_ticks():
                tick.label.set_fontsize(22) 
        
ax.tick_params('both', length=5, width=2, which='major')
ax.tick_params('both', length=3, width=1, which='minor')
    
#plt.legend(loc='upper center', fontsize=20)
plt.legend(bbox_to_anchor=(0.4, 1), fontsize=20)
fig.set_tight_layout(True)

a = plt.axes([.60, .5, .32, .32])
plt.plot(ValueFilePooled, Option2RatePooled, 'or', markersize=12, 
         ls = 'solid', lw = 3.5, label=r'$\Delta d = 0$')
plt.plot(ValueFilePooled2, Option2RatePooled2, 'sg', markersize=12, 
         ls = 'solid', lw = 3.5, label=r'$\Delta d = 2$')
plt.plot(ValueFilePooled3, Option2RatePooled3, '^b', markersize=12, 
         ls = 'solid', lw = 3.5, label=r'$\rho_d = 2$')


plt.xlabel(r'mean deficit $d_m$', fontsize=14)
plt.ylabel(r'rate option "drink" chosen', fontsize=14)

#plt.xlim(5,10)
#plt.ylim(0,1)

a.xaxis.set_major_locator(ticker.MultipleLocator(1))
a.xaxis.set_minor_locator(ticker.MultipleLocator(0.5))
a.yaxis.set_major_locator(ticker.MultipleLocator(0.1))
a.yaxis.set_minor_locator(ticker.MultipleLocator(0.05))
for axis in ['top','bottom','left','right']:
    a.spines[axis].set_linewidth(2)
    
for tick in a.xaxis.get_major_ticks():
                tick.label.set_fontsize(14) 
for tick in a.yaxis.get_major_ticks():
                tick.label.set_fontsize(14) 
        
a.tick_params('both', length=5, width=2, which='major')
a.tick_params('both', length=3, width=1, which='minor')
    
#plt.legend(loc='center', fontsize=12)
fig.set_tight_layout(True)
```
